# Relating the variable with Scaterplot

# #Here we relating data of total Cases and Death Cases

```python
#scatterplot :- A scatter plot is a diagram where each value in the data set
#is represented by a dot.
#We have one input variable means independent variable and one output variable
#means dependent variable .
#input means X and output means Y
```

```python
X=data["Total Cases"]
Y=data["Deaths"]
plt.scatter(X,Y,color='red',marker='*',s=40)
plt.xlabel("Total Cases",color="red") # xlabel() defines the label of x-axis 
plt.ylabel("Total Death cases",color="green") #ylabel() defines the label of x-axis
plt.title("Total Cases Vs Death Cases",color="Blue") #title() is used to give title of this scatter plot
plt.show()

```

# Here we relating data of total Cases and Total Active Cases

```python
X=data["Total Cases"] 
Y=data["Active"]
plt.scatter(X,Y)
plt.xlabel("Total Cases") # xlabel() defines the label of x-axis
plt.ylabel("Total Active cases") #ylabel() defines the label of x-axis
plt.show()
```

# Here we relating data of Active Cases and Discharge Cases


```python
X=data["Active"] 
Y=data["Discharged"]
plt.scatter(X,Y,color="red",marker="^",s=60) #here we plot graph by using ^ marker and its s is the size 
plt.xlabel("Active Cases",color="green")# xlabel() defines the label of x-axis 
plt.ylabel("Discharge Cases",color="green")# ylabel() defines the label of y-axis 
plt.title("Active cases vs Discharged Cases",color="blue") #title() is used to give title of this scatter plot
plt.show()

```

# Bar plot


A bar plot or bar chart is a graph that represents the category of data with rectangular bars with lengths and heights that is proportional to the values which they represent. The bar plots can be plotted horizontally or vertically. A bar chart describes the comparisons between the discrete categories

```python
#Here we show the representation of bar plot between State Wise and total case
plt.bar(data["State/UTs"],data["Total Cases"],color=["maroon","blue"])
plt.xlabel("States of India",color="green")# xlabel() defines the label of x-axis 
plt.ylabel("Total Cases in India",color="maroon")# ylabel() defines the label of y-axis 
plt.title("Total Cases Vs State/UTs",color="blue") #title() is used to give title of this bar plot
plt.xticks(rotation=90)
plt.show()
```

# countplot() 


countplot() method is used to Show the counts of observations in each categorical bin using bars. Parameters : This method is accepting the following parameters that are described below: x, y: This parameter take names of variables in data or vector data, optional, Inputs for plotting long-form data

```python
#here we countplot of Discharged Ratio of india
sns.countplot(data=data,x="Discharge Ratio (%)") #here we countplot of dischared ratio 
c=data["Discharge Ratio (%)"].value_counts() # we count the discharge ratio of value
plt.yticks(c)
plt.xticks(rotation=90)
plt.show()
```

```python
#here we countplot of Deaths of Indian people And take hue ofDeath ratio(%) cases of india
plt.figure(figsize=(9,9))
sns.countplot(data=data,x="Deaths",hue="Death Ratio (%)",color="red")
plt.xlabel("Deaths",color="blue")
plt.ylabel("Count",color="pink")
plt.xticks(rotation=90)
plt.show()
```

# pie plot

A Pie Chart is a circular statistical plot that can display only one series of data. The area of the chart is the total percentage of the given data. The area of slices of the pie represents the percentage of the parts of the data. The slices of pie are called wedges.

```python
#pie graph of death 
data["Deaths"].value_counts().plot(kind="pie",autopct="%.01f%%")
plt.show()
```

```python
#pie graph of States 
data["State/UTs"].value_counts().plot(kind="pie",autopct="%.01f%%")
plt.show()
```

# Box Plot

A Box Plot is also known as Whisker plot is created to display the summary of the set of data values having properties like minimum, first quartile, median, third quartile and maximum.

```python
X=data["Total Cases"] #here we show the box plot of Total cases 
plt.boxplot(X) # it is a method to represent the box plot
plt.show()
```
